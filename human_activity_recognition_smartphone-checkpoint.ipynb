# Human Activity Recognition
Human Activity Recognition database built from the recordings of 30 subjects performing __activities of daily living (ADL)__ while carrying a waist-mounted smartphone with embedded inertial sensors.

_Data Set Information:_
The experiments have been carried out with a group of 30 volunteers within an age bracket of 19-48 years. Each person performed six activities __(WALKING, WALKING_UPSTAIRS, WALKING_DOWNSTAIRS, SITTING, STANDING, LAYING)__ wearing a smartphone (Samsung Galaxy S II) on the waist. Using its embedded accelerometer and gyroscope, we captured 3-axial linear acceleration and 3-axial angular velocity at a constant rate of 50Hz. The experiments have been video-recorded to label the data manually. 

The sensor signals (accelerometer and gyroscope) were pre-processed by applying noise filters and then sampled in fixed-width sliding windows of 2.56 sec and 50% overlap (128 readings/window). The sensor acceleration signal, which has gravitational and body motion components, was separated using a Butterworth low-pass filter into body acceleration and gravity. The gravitational force is assumed to have only low frequency components, therefore a filter with 0.3 Hz cutoff frequency was used. From each window, a vector of features was obtained by calculating variables from the time and frequency domain.

### Reducing dimensions to visualize better
Now since dimesions are too many here, we need to reduce it to visualize plots and gain insights.

We would use __Principal Component Analysis__ for the same.

```python
# Create datasets
trans_data = data_.copy()
activity_data = trans_data.pop('activity')
print(f'Feature Data Shape: {trans_data.shape}')
print(f'Activity Label Data Shape: {activity_data.shape}')

# Scale data
scl = StandardScaler()
trans_data = scl.fit_transform(trans_data)

# Reduce dimensions
pca = PCA(n_components=0.9, random_state=0)
trans_data = pca.fit_transform(trans_data)
print(f'Feature Data Shape(After Reduction): {trans_data.shape}')
```

```python
activity_count = activity_data.value_counts()
activity_count
```

```python
### Plot Activities
plt.figure(figsize=(15,10))

# Get colors
n = activity_data.unique().shape[0]
colormap = get_cmap('viridis')
colors = [rgb2hex(colormap(col)) for col in np.arange(0, 1.01, 1/(n-1))]
rng = np.random.RandomState(0)
sizes = 500 * rng.rand(n)
print(sizes)

# Plot each activity
for i, activity in enumerate(activity_count.index):
    # Select all rows with current activity
    mask = (activity_data==activity).values
    plt.scatter(x=trans_data[mask][:,0], y=trans_data[mask][:,1], c=colors[i], alpha=0.5, label=activity, s=sizes)
plt.title('PCA: Activity Visualisation')
plt.legend()
```

As you can see, __the activities are not seperable entirely__. This is because we still have 64 dimensions and PCA doesn't work very well with non-linear data. Lets' try __manifold learning__ to reduce the dimensions further. 

```python
## Manifold Learing
# We choose TSNE(t-distributed stochastic neibhour embedding) as the main purpose of t-SNE is visualization of high-dimensional data. Hence, it works best when the data will be embedded on two or three dimensions.
tsne = TSNE(random_state=0)
tsne_transformed = tsne.fit_transform(trans_data)
print(f'Feature Data Shape(After Transformation): {tsne_transformed.shape}')
```

```python
plt.figure(figsize=(15,10))
for i, activity in enumerate(activity_count.index):
    # Select all rows with current activity
    mask = (activity_data==activity).values
    plt.scatter(x=tsne_transformed[mask][:,0], y=tsne_transformed[mask][:,1], c=colors[i], alpha=0.5, label=activity, s=sizes)
plt.title('TSNE: Activity Visualisation')
plt.legend()
```

__From above plot, we see activities are mostly separable__.

### Plot Histograms Per Activity
We are interested in discriminating between activities based on activity data.

Check the distribution of data between different activities. 

```python
fig, axarr = plt.subplots(6, 1, figsize=(10,25))

# Assign colors for each component and the names
hist_colors = ['#E69F00', '#56B4E9']
names = ['Component 1','Component 2']

for i, activity in enumerate(activity_count.index):
    hist_mask = (activity_data==activity).values
    x1 = tsne_transformed[hist_mask][:,0]
    x2 = tsne_transformed[hist_mask][:,1]
    axarr[i].hist([x1, x2], normed=True,color = hist_colors, label=names, stacked=True)
    title = activity + ' Measurement Distribution'
    axarr[i].set_ylabel(activity)
    axarr[i].set_xlabel('Measurement Components')
    axarr[i].set_title(title)
    axarr[i].legend()
    
# Make the chart fill out the figure better.
fig.tight_layout()
plt.show()
```

__As observed from above plots, most of the data fallow the gaussian distribution__

## Classification
In this section we would apply different classification algorithms and compare the results.

### Prepare the data set
In this section we would verify the dimensional reduction done and prepare training and testing dataset.

```python
features = data_.iloc[:,1:]
features.head()
```

```python
features = scl.fit_transform(features)
checkpca = PCA().fit(features)
plt.plot(np.cumsum(checkpca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative Explained Variance');
```

__Above proves that 64 components covers 90% of the variance__. Hence, we would use:
* 'activity_data' as label data.
* 'trans_data' as feature data.


```python
#Split training testing data
enc = LabelEncoder()
label_encoded = enc.fit_transform(activity_data)
X_train,X_test, y_train, y_test = train_test_split(trans_data, label_encoded, random_state=0)
print(f'Training data, features: {X_train.shape}, labels: {y_train.shape}')
print(f'Testing data, features: {X_test.shape}, labels: {y_test.shape}')
```

### Support Vector Machine Classifier
These are powerful and flexible class of supervised algorithms for classification.

We will use Grid Search CV for hypertuning the paramters.

```python
from sklearn.svm import SVC
from pprint import pprint

# C
C = [.0001, .001, .01]

# gamma
gamma = [.0001, .001, .01, .1, 1, 10, 100]

# degree
degree = [1, 2, 3, 4, 5]

# kernel
kernel = ['linear', 'rbf', 'poly']

# probability
probability = [True]

# Create the random grid
random_grid = {'C': C,
              'kernel': kernel,
              'gamma': gamma,
              'degree': degree,
              'probability': probability
             }
pprint(random_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV

# First create the base model to tune
svc = SVC(random_state=0)

# Definition of the random search
random_search = RandomizedSearchCV(estimator=svc,
                                   param_distributions=random_grid,
                                   n_iter=50,
                                   scoring='accuracy',
                                   cv=3, 
                                   verbose=1, 
                                   random_state=0)

# Fit the random search model
random_search.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit

# Create the parameter grid based on the results of random search 
C = [1, 3, 5, 10, 50]
degree = [2, 3, 4, 5]
gamma = [0.0001, 0.0005, 0.001, 0.005]
probability = [True]

param_grid = [
  {'C': C, 'kernel':['poly'], 'degree':degree, 'probability':probability},
  {'C': C, 'kernel':['rbf'], 'gamma':gamma, 'probability':probability}
]

# Create a base model
svc = SVC(random_state=0)

# Manually create the splits in CV in order to be able to fix a random_state (GridSearchCV doesn't have that argument)
cv_sets = ShuffleSplit(n_splits = 3, test_size = .33, random_state = 0)

# Instantiate the grid search model
grid_search = GridSearchCV(estimator=svc, 
                           param_grid=param_grid,
                           scoring='accuracy',
                           cv=cv_sets,
                           verbose=1)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
print("The best hyperparameters from Grid Search are:")
print(grid_search.best_params_)
print("")
print("The mean accuracy of a model with these hyperparameters is:")
print(grid_search.best_score_)
```

```python
best_svc = grid_search.best_estimator_
best_svc.fit(X_train, y_train)
svc_pred = best_svc.predict(X_test)
```

```python
# Accuracy Score
from sklearn.metrics import accuracy_score
print("The test accuracy is: ")
print(accuracy_score(y_test, svc_pred))
print("The training accuracy is: ")
print(accuracy_score(y_train, best_svc.predict(X_train)))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, svc_pred, target_names=activity_count.index))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

SVC_mat = confusion_matrix(y_test, svc_pred)
sns.heatmap(SVC_mat, square=True, annot=True, fmt='d', cbar=False, xticklabels=activity_count.index, yticklabels=activity_count.index)
plt.xlabel('True Label')
plt.ylabel('Predicted Label')
plt.title('SVM: Confusion Matrix')
```

### Logistic Regression Classifier
This is Logistic Regression algorithm for multi-class classification. It uses Softmax Function for classification. It calculates probablity of a label considering the probabality among all labels.

```python
from sklearn.linear_model import LogisticRegression

# C
LRC = [float(x) for x in np.linspace(start = 0.1, stop = 1, num = 10)]

# multi_class
LR_multi_class = ['multinomial']

# solver
solver = ['newton-cg', 'sag', 'saga', 'lbfgs']
 
# class_weight
class_weight = ['balanced', None]

# penalty
penalty = ['l2']

# Create the random grid
random_grid = {'C': LRC,
               'multi_class': LR_multi_class,
               'solver': solver,
               'class_weight': class_weight,
               'penalty': penalty}

log_reg = LogisticRegression(random_state=0)

# Definition of the random search
lr_random_search = RandomizedSearchCV(estimator=log_reg,
                                   param_distributions=random_grid,
                                   n_iter=50,
                                   scoring='accuracy',
                                   cv=3, 
                                   verbose=1, 
                                   random_state=0)

# Fit the random search model
lr_random_search.fit(X_train, y_train)
```

```python
print("The best hyperparameters from Random Search are:")
print(lr_random_search.best_params_)
print("")
print("The mean accuracy of a model with these hyperparameters is:")
print(lr_random_search.best_score_)
```

```python
LRC = [float(x) for x in np.linspace(start = 0.1, stop = 1, num = 10)]
lr_multi_class = ['multinomial']
solver = ['sag', 'lbfgs']
class_weight = ['balanced']
penalty = ['l2']

param_grid = {'C': LRC,
               'multi_class': lr_multi_class,
               'solver': solver,
               'class_weight': class_weight,
               'penalty': penalty}

# Create a base model
log_reg = LogisticRegression(random_state=0)

# Manually create the splits in CV in order to be able to fix a random_state (GridSearchCV doesn't have that argument)
cv_sets = ShuffleSplit(n_splits = 3, test_size = .33, random_state = 8)

# Instantiate the grid search model
lr_grid_search = GridSearchCV(estimator=log_reg, 
                           param_grid=param_grid,
                           scoring='accuracy',
                           cv=cv_sets,
                           verbose=1)

# Fit the grid search to the data
lr_grid_search.fit(X_train, y_train)
```

```python
print("The best hyperparameters from Grid Search are:")
print(lr_grid_search.best_params_)
print("")
print("The mean accuracy of a model with these hyperparameters is:")
print(lr_grid_search.best_score_)
```

```python
best_log_reg = lr_grid_search.best_estimator_
best_log_reg.fit(X_train, y_train)
log_reg_pred = best_log_reg.predict(X_test)
```

```python
# Accuracy Score
print("The test accuracy is: ")
print(accuracy_score(y_test, log_reg_pred))
print("The training accuracy is: ")
print(accuracy_score(y_train, best_log_reg.predict(X_train)))
```
